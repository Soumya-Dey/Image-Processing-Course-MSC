```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data1 = plt.imread("/content/drive/MyDrive/#2-M.Sc Computer Science [SXC]/1st Year/Books and notes/Sem 4/Image Processing/SP+AN/AN_Image/python.jpg")
data1 = plt.imread(r"D:\#College\M.Sc Computer Science [SXC]\1st Year\Books and notes\Sem 4\Image Processing\SP+AN\AN_Image\python.jpg")
```

```python
shape1 = data1.shape
```

```python
r,c,d = shape1[0], shape1[1], shape1[2]
R,G,B = data1[:,:,0], data1[:,:,1], data1[:,:,2]

gray1 = .2989*R + .5870*G + .114*B
```

```python
shape2 = gray1.shape
```

```python
r1, c1 = shape2[0], shape2[1]
```

```python
plt.axis('off')
plt.title('RGB Image')
plt.imshow(data1)
plt.show()
```

```python
plt.axis('off')
plt.title('GrayScale Image')
plt.imshow(gray1, cmap='gray')
plt.show()
```

```python
gray2 = np.copy(gray1)
for i in range(gray1.shape[0]):
    for j in range(gray1.shape[1]):
        gray2[i, j] = gray1[i, gray1.shape[1] - j - 1]
```

```python
plt.axis('off')
plt.title('Mirror Image')
plt.imshow(gray2, cmap='gray')
plt.show()
```

```python
gray3 = np.copy(gray1)
for i in range(gray1.shape[0]):
    for j in range(gray1.shape[1]):
        gray3[i, j] = gray1[gray1.shape[0] - i - 1, j]
```

```python
plt.axis('off')
plt.title('Mirror Image')
plt.imshow(gray3, cmap='gray')
plt.show()
```

```python
gray4 = np.copy(gray1)
for i in range(gray1.shape[0]):
    for j in range(gray1.shape[1]):
        gray4[i, j] = gray1[gray1.shape[0] - i - 1, gray1.shape[1] - j - 1]
```

```python
plt.axis('off')
plt.title('Mirror Image')
plt.imshow(gray4, cmap='gray')
plt.show()
```

```python
plt.axis('off')
plt.title('Transpose Image')
plt.imshow(gray1.T, cmap='gray')
plt.show()
```

```python
data2 = np.copy(data1)
for i in range(0, r):
  for j in range(0, c//2):
    for k in range(0, d):
      t = data2[i, j, k]
      data2[i, j, k] = data2[i, c-1-j, k]
      data2[i, c-1-j, k]=t
```

```python
plt.axis('off')
plt.title('Mirror Image')
plt.imshow(data2, cmap='gray')
plt.show()
```

```python
width, height, C = data1.shape
data_flip = np.zeros((width, height, C), dtype=np.uint8)
```

```python
for i,row in enumerate(data1):
    data_flip[width - 1 - i, :, :] = row
```

```python
plt.axis('off')
plt.title('Flip Image')
plt.imshow(data_flip, cmap='gray')
plt.show()
```
