```python
#to display the rgb image
img1=r'/content/drive/MyDrive/#2-M.Sc Computer Science [SXC]/1st Year/Books and notes/Sem 4/Image Processing/SP+AN/image folder/storm.jpg'
data1=plt.imread(img1)
R,G,B=data1[:,:,0],data1[:,:,1],data1[:,:,2]
fig=plt.figure(figsize=(15,15))
rows=2
columns=4


fig.add_subplot(rows,columns,1)
plt.title('RGB Image')
plt.imshow(data1)

#to display the GRB channel
data2=data1.copy()
data2[:,:,0]=G
data2[:,:,1]=R
data2[:,:,2]=B
fig.add_subplot(rows,columns,2)
plt.title('GRB channel')
plt.imshow(data2)

#to display the RBG channel
data3=data1.copy()
data3[:,:,0]=R
data3[:,:,1]=B
data3[:,:,2]=G
fig.add_subplot(rows,columns,3)
plt.title('RBG channel')
plt.imshow(data3)

#to display the BGR channel
data4=data1.copy()
data4[:,:,0]=B
data4[:,:,1]=G
data4[:,:,2]=R
fig.add_subplot(rows,columns,4)
plt.title('BGR channel')
plt.imshow(data4)

#to display the RGR channel
data5=data1.copy()
data5[:,:,0]=R
data5[:,:,1]=G
data5[:,:,2]=R
fig.add_subplot(rows,columns,5)
plt.title('RGR channel')
plt.imshow(data5)

plt.show()
```
