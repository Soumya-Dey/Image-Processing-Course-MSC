#### 2. Get the count of files of each extension in a folder

```python
def getGropuedObj(path):
    files = getFiles(path)
    groupedObj = {}

    for f in files:
        ext = f.split(".")[-1]
        groupedObj[ext] = groupedObj[ext] + 1 if ext in groupedObj else 1
        
    return groupedObj
```

#### move files

```python
import shutil
```

```python
def moveFiles(src, dest):
    for f in os.listdir(src):
        shutil.move(os.path.join(src, f), os.path.join(dest, f))
```

```python
moveFiles(path, basePath)
```

#### copy files

```python
def copyFiles(src, dest):
    for f in os.listdir(src):
        shutil.copy(os.path.join(src, f), os.path.join(dest, f))
```

```python
moveFiles(basePath, path)
```
