- Write a function in Python (WAFP) to access the image folder and return the count of the total number of files
- WAFP to count the number of files of each extension (.png, .jpg, .bmp, etc.) type and return the values

```python
'''
WAFP to access the image folder and return the count of the total number of files
'''
def countOfFiles(path):
    return len(os.listdir(path))
```

```python
p = "D://ML & DL//IP@SP//image folder"
print("Number of files present in the given directory is",countOfFiles(p))
```

```python
'''
WAFP to count the number of files of each extension (.png, .jpg, .bmp, etc.) type and return the values
'''
def countFilesOfEachType(path):
    d = dict()
    list_of_files = os.listdir(path)
    for file in list_of_files:
        extension = file.split(".")[1]
        if extension in list(d.keys()):
            d[extension] += 1
        else:
            d[extension] = 1
    return d
```

```python
p = "D://ML & DL//IP@SP//image folder"
eachType = countFilesOfEachType(p)
print("Number of files of each type is",countFilesOfEachType(p))
```

### Moving image files

- Move file from image folder to new_directory

```python
import shutil
```

```python
def move_files(src_path, dest_path):
    src_files = os.listdir(src_path)
    for file in src_files:
        source = os.path.join(src_path, file)
        destination = os.path.join(dest_path, file)
        shutil.move(source, destination)
```

```python
src_path = "D://ML & DL//IP@SP//image folder"
dst_path = "D://ML & DL//IP@SP//new_directory"
```

```python
move_files(src_path, dst_path)
```

- Save a copy of the task 1 folder
- WAFP to rename the files in the task 1 folder using the following convention:
    1_MIR, 2_MIR, 1_TIR1, 2_TIR2, etc.
- WAFP to create a directory called storm_images, under which you need to create 3 sub-directories TIR1, TIR2 and MIR
- WAFP to transfer the renamed files to the corresponding directories TIR1, TIR2 and MIR

```python
def renameFiles(path):
    list_of_files = os.listdir(path)
    cnt_MIR = 0
    cnt_TIR1 = 0
    cnt_TIR2 = 0
    for file in list_of_files:
        type_of_file = file.split(".")[0]
        if type_of_file.endswith("MIR"):
            cnt_MIR += 1
            s = os.path.join(path,file)
            d = os.path.join(path,str(cnt_MIR)+"_"+"MIR"+".jpg")
            os.rename(s,d)
        elif type_of_file.endswith("TIR1"):
            cnt_TIR1 += 1
            s = os.path.join(path,file)
            d = os.path.join(path,str(cnt_TIR1)+"_"+"TIR1"+".jpg")
            os.rename(s,d)
        elif type_of_file.endswith("TIR2"):
            cnt_TIR2 += 1
            s = os.path.join(path,file)
            d = os.path.join(path,str(cnt_TIR2)+"_"+"TIR2"+".jpg")
            os.rename(s,d)
```

```python
p = r"D:\ML & DL\IP@SP\task 1"
renameFiles(p)
```

```python
def mkdirsubdir(path):
    parent_dir = os.path.join(path,"storm_images")
    os.mkdir(parent_dir)
    subdir_TIR1 = os.path.join(parent_dir,"TIR1")
    os.mkdir(subdir_TIR1)
    subdir_TIR2 = os.path.join(parent_dir,"TIR2")
    os.mkdir(subdir_TIR2)
    subdir_MIR = os.path.join(parent_dir,"MIR")
    os.mkdir(subdir_MIR)
```

```python
p = r"D:\ML & DL\IP@SP\task 1"
mkdirsubdir(p)
```

```python
def filetransfer(src_path, mir_path, tir1_path, tir2_path):
    list_of_files = os.listdir(src_path)
    for file in list_of_files:
        if file.split(".")[0].endswith("TIR1"):
            s = os.path.join(src_path,file)
            d = os.path.join(tir1_path,file)
            shutil.move(s, d)
        elif file.split(".")[0].endswith("TIR2"):
            s = os.path.join(src_path,file)
            d = os.path.join(tir2_path,file)
            shutil.move(s, d)
        if file.split(".")[0].endswith("MIR"):
            s = os.path.join(src_path,file)
            d = os.path.join(mir_path,file)
            shutil.move(s, d)
```

```python
s = r"D:\ML & DL\IP@SP\task 1"
m = r"D:\ML & DL\IP@SP\task 1\storm_images\MIR"
t1 = r"D:\ML & DL\IP@SP\task 1\storm_images\TIR1"
t2 = r"D:\ML & DL\IP@SP\task 1\storm_images\TIR2"
filetransfer(s,m,t1,t2)
```
