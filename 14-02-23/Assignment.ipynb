```python
import os
from skimage import io, img_as_float
import cv2
```

#### 1. Read all images from "image folder" and print the shape

```python
def getSizes(path):
  images = os.listdir(path)
  for imgPath in images:
    print(imgPath) if imgPath.endswith("db") else print(imgPath, io.imread(os.path.join(path, imgPath)).shape)
```

```python
path = "/content/drive/MyDrive/#2-M.Sc Computer Science [SXC]/1st Year/Books and notes/Sem 4/Image Processing/SP+AN/image folder/"
getSizes(path)
```

#### 2. Normalise all imgaes of the "image folder"

```python
def normaliseImages(path, pathToCopy):
  os.mkdir(pathToCopy)

  images = os.listdir(path)
  for imgPath in images:
    try:
      if imgPath.endswith("db"):
        print(imgPath)
      else: 
        img = io.imread(os.path.join(path, imgPath))
        imgFloat = img_as_float(img)
        cv2.imwrite(os.path.join(pathToCopy, imgPath), imgFloat)
    except (IOError,ValueError) as e:
        print('could not read the', imgPath ,':',e,'hence skipping it.')
```

```python
pathToCopy = "/content/drive/MyDrive/#2-M.Sc Computer Science [SXC]/1st Year/Books and notes/Sem 4/Image Processing/SP+AN/14-02-23/Normalised"
normaliseImages(path, pathToCopy)
```
