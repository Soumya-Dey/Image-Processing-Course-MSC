```python
img1 = plt.imread("/content/drive/MyDrive/#2-M.Sc Computer Science [SXC]/1st Year/Books and notes/Sem 4/Image Processing/SP+AN/Barca Images/FC_Barcelona_Team_2,_2011.jpg")
R, G, B= img1[:, :, 0], img1[:, :, 1], img1[:, :, 2]
gray = .2989*R + .5870*G + .114*B
```

```python
plt.imshow(gray.T, cmap="gray")
```

```python
bw = gray
for i in range(gray.shape[0]):
  for j in range(gray.shape[1]):
    bw[i][j] = 255 if gray[i][j] > 127 else 0

plt.imshow(bw, cmap="gray")
```
